```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
import seaborn as sns
HOST = '127.0.0.1'
USER = 'root'
PASS = '1234'
DB   = 'classicmodels'


# CODE IS RUNNING PERFECTLY,
# BUT IT MAY BE A LITTLE BIT MORE PROFESSIONAL !

#: Create the connection
cnx = mysql.connector.connect(user=USER, password=PASS, host=HOST, database=DB)
cnx.autocommit = True

#: Create the cursor
mycursor = cnx.cursor()
```

```python
df = pd.read_sql('SELECT creditLimit, quantityOrdered,priceEach,amount, buyPrice, quantityInStock, MSRP FROM customer_payment_product_orders', con=cnx)

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Set plot title 
plt.title('Correlation Matrix Heatmap')

# Show plot
plt.show()
```

```python

# Read the data from SQL into a pandas dataframe
df = pd.read_sql('SELECT quantityOrdered, customerName, creditLimit FROM customer_payment_product_orders', con=cnx)

# Create the histogram with customer names as hue
sns.histplot(data=df, x="creditLimit", hue="customerName")

# Add x and y labels
plt.xlabel("Credit Limit")
plt.ylabel("Count")

# Show the plot
plt.show()
```
